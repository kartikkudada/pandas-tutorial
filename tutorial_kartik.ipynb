```python
bios['height_category'] =bios['height_cm'].apply(lambda height:'short' if height < 165 else ('average' if height < 185 else 'tall'))
```

```python
bios.head()
```

```python
def categorize_athelete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 85:
        return 'Middleweight'
    else:
        return 'Heavyweight'
    
```

```python
bios['category'] = bios.apply(categorize_athelete,axis=1)
```
